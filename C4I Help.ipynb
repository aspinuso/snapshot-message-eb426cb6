# Welcome to Climat4Impact v2 *(alpha)*


This new version of the C4I portal is a complete redesing of the old website (still available at http://climate4impact.eu). 

We addressed in v2 several improvements to the search and data analysis functionalities, aiming at supporting climate change impact researchers with more flexibility in the way they access and manipulate the data provided by the Earth System Grid Federation (ESGF). 

C4I v2 integrates with the new ESGF authentication and authorisation infrastructure (AAI). This service is still under development, so we apologie in advance for downtimes or instabilty of the login & authorisation functionality. 

In this short guide we explain some of the new features.

## How do I serach for data?

In the ***Data Discovery Page*** you can first select the project you want to access the data from (CMIP6 - CMIP5 - CORDEX).

**NOTE:** at the time of writing this guide the authorisation to access CORDEX data is not in place yet. We will enable this option as soon as the underlying AAI will be upgraded.

Once you have selected the project the portal offers the possibility to compose queries interactively by combining different properties such as: *parameters*, *frequency*, *experiment*, *model*, *model member*.


<img src="images/search.png" alt="search" style="height:350px;width:800px;"/>


By performing a selection across the different properties, the search options will adapt to the current choices. By default, to facilitate identifying the most relevant options, a subset of the possible values will be presented grouped by different topics. This is the *Scoped* view. However, you can always change this presentation by switching the knob on the bottom left of the screen, from *Scoped* to *Extended*. The *Extended* view will display, for each property, all the possible options available in the current query configuration. 

In the image above you can see how the summary of the current query is always displayed on the top of the screen. These can be interactively removed from the query, resulting in the live update of the options available in each property.

## How do I select and download the data?

After you have prepared your query, C4I offers you more ways of accessing the data.

By pressing the ***Download*** button at the bottomo of the page, C4I will present the list of datasets available in the ESGF that are matching your query. The list will be ordered by ***mirror***, so that you can easily find and select the data of interest from the same ESGF node. If needed, you can change the default ordering interactively.

<img src="images/select.png" alt="select" style="height:350px;width:800px;"/>

Each dataset in the list is associated with a clickable DOI (right side the page). The link will will open the page which describe the metadata of the related model and experiment.

By clcking on the ***Link List*** button, a file containing the list of URLs associated with the data can be downloaded. The file is reprented in the standard metalink format (https://en.wikipedia.org/wiki/Metalink), so that it can used in your processing scripts adopting generic libraries or clients already available to download the data independently, and at the desired location.

### How to prepare the data for analyses in C4I?

The ***Download*** page offers you the opportunity to access the new experimental analysis  functionalities of C4I. By clicking on ***Export To Notebook*** you can request the portal to automatically export and manage the data onto a *JupyterLab* notebook (https://jupyterlab.readthedocs.io), which will be made available for you to use. 

C4I will prepare the resources needed to provide you with a notebook enviroenment. This will contain the selected data (within the limit of the available storage), a collection of libraries to be used programmatically (eg. *icclim, pandas, xarray*) and a set of tools which will allow you to **update**, **restore**, **share** and **reproduce** the progress of your research. The environment is provided thanks to the SWIRRL API (Software for Reproducible Research Lab). You will find a guide on how to use SWIRRL![enter image description here](https://assets.gitlab-static.net/uploads/-/system/project/avatar/17988184/Screenshot_2020-04-09_at_10.37.38.png?width=32) in the notebook (Welcome to SWIRRL file).


<img src="images/swirrl.png" alt="swirrl" style="height:350px;width:800px;"/>



**NOTE** Your *Jupyter Lab* instance could be accessible before all the data is copied into your storage. Please refer to the SWIRRL guide to read more about how the data is staged and and managed into your notebook.


### Are there examples to learn how to implement new analyses? 

To help with developing your own first analysis notebook, we have collected a number of examples. You can browse and download those you are interested in from the IS-ENES3 GitLab repository (https://gitlab.com/is-enes-cdi-c4i/notebooks). By selecting in C4I the data mentioned in the  sample notebooks, you can  export the files to your existing notebook or ask C4I to create a new one.  Upload the sample notebooks to your Jupyter Lab and, once the data has been copied (see the Welcome to SWIRRL guide for details), you will be able to test, change and experiment youself with this new experimental facility!





### Can I reduce the amount of data downloaded by subsetting?

Yes, this is possible. However not all of the ESFG nodes are supporting this type of service.
By clickin on the ***Subsetting*** button of the ***Data Disovery*** page, you can choose the datasets your are interetesed that match you query paramenters. By pressing the ***send to process*** button C4I will show you a configuration panel where you can set your subsetting parementes. This feature is currently linked to the experimental notebook facility, so the data will be made availabe for you in your Jupyter Lab instance. 

<img src="images/subsetting.png" alt="subsetting" style="height:350px;width:800px;"/>

In the IS-ENES project, we are working on implementing a more flexible and performing technology to enable subsetting and other type of common data transformation operation. This will make the data smaller and more usable for you to proceed with futher analysis. We will in tbe made available for you in the coming releases!
